## Matplotlib and Seaborn P

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# set default seaborn style
sns.set_style(
    'ticks',
    rc={
        'axes.spines.right': False,
        'axes.spines.top': False,
        'patch.edgecolor': 'k'
    }
)
```

```python
sns.axes_style()
```

```python
# histogram

# set figure size
plt.figure(figsize=(12,6))

sns.distplot(df['total_bill'],kde=False, color='#ff8800')

sns.despine(offset=10, trim=True)
plt.show()
```

```python
# histogram w/ KDE

# set figure size
plt.figure(figsize=(12,6))

sns.distplot(df['total_bill'], color='#ff8800')

sns.despine(offset=10, trim=True)
plt.show()
```

```python
# KDE

# set figure size
plt.figure(figsize=(12,6))

# Plot a filled kernel density estimate
sns.distplot(
    df['total_bill'],
    hist=False, 
    color='#ff8800', 
    kde_kws={'shade': True}
)
sns.despine(trim=True)
plt.show()
```

```python
# boxplot
sns.boxplot(df['total_bill'], color='#ff8800')
```
