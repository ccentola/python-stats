## Tables, Boxplots, and Histograms in Python

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = sns.load_dataset('tips')
```

```python
sns.set(rc={
    'figure.figsize':(8,4),
    'patch.force_edgecolor':True
})

sns.set_style(
    'white', 
    {
        'axes.spines.right': False,
        'axes.spines.top': False,
        'figure.figsize':(12,6)
    }
)

```

```python
# look at first few rows
data.head()
```

```python
# summary statistics
data.describe()
```

```python
# creating a histogram
sns.distplot(
    data['total_bill'], 
    kde=False,color='blue',
    hist_kws=dict(edgecolor='k', linewidth=2)
).set_title('Dist. of Total Bill')

plt.show()
```

```python
# creating a histogram
sns.distplot(
    data['tip'], 
    kde=True,
    color='green',
    hist_kws=dict(edgecolor='k', linewidth=2)
).set_title('Dist. of Tips')

plt.show()
```

```python
# total bill boxplot
sns.boxplot(data['total_bill'])
plt.show()
```

```python
# tips boxplot
sns.boxplot(data['tip'])
plt.show()
```

```python
# boxplots by group
sns.boxplot(data['tip'],data['smoker'])
plt.show()
```

```python
# compare hists
g = sns.FacetGrid(data,row='smoker')
g = g.map(plt.hist,'tip')

plt.show()
```

```python
# boxplot of tips by meal
sns.boxplot(data['tip'],data['time'])

g = sns.FacetGrid(data,row='time')
g = g.map(plt.hist,'tip')


plt.show()
```

```python
sns.boxplot(data['tip'],data['day'])

g = sns.FacetGrid(data,row='day')
g = g.map(plt.hist,'tip')

plt.show()
```
