# Inferential Statistics

**inference** - a method of making judgement about the unknown, drawing on what is already known to be true.

## Inferential vs. Descriptive

* use descriptive statistics if the cases you are studying represent the entire population.
* use inferential statistics when the cases you are studying do not represent the entire population of intereset and you wish to generalize.

## Probability Distributions

* **continious** - can take any value withina specified range
* **discrete** - can only take certian values

### Normal Distribution

* standard normal distribution or Z distribution: $\mu = 0$, $\sigma = 1$ 
* all normal distribution sregardless of their mean and standard deviation share the following characteristics:
    * symmetry
    * unimodality (a single most common value)
    * a continious range of $-\infty$ to $+\infty$
    * a total area under the curve of 1
    * a common value for the mean, median, and mode
* **emperical rule** - all normal distrubutions have a similar shape so we can assume:
    * about 68% of the data will fall within one standard deviation of the mean
    * about 95% of the data will fall within two standard deviations of the mean
    * about 99% of the data will fall within three standard deviations of the mean
* **Z-score** - distance of a data point from the mean
$$Z=\frac{x - \mu}{\sigma}$$

### Binomial Distribution

* discrete
* events are generated by a *Bernoulli process*. the distribution describes the number of successes in n trails.
* data represented by the binomial distribution must meet four requirements:
    1. the outcome of each trial is one of two mutually exclusive outcomes
    2. each trial is independent
    3. the probability of success, denoted as $p$, is constant for every trial
    4. there is a fixed number of trials

$${n \choose k}p^k(1-p)^{n-k}$$
$${n \choose k} = nCk = \frac{n!}{k!(n-k)!}$$

* if both $np$ and $n(1-p)$ are greater than or equal to 5, the binomial distribution may be approximated by the normal distribution

## Independent and Dependent Variables

* **dependent** - represent the outcome of the study.
* **independent** - presumed to have some influence over the value of the dependent variable(s).
* **control** - might influence the dependent variable, but are not the main focus.

**Ordinary Least Squares (OLS)**
$$
Y = \beta_0 + \beta_{1}X_{1} + \beta_{2}X_{2} + \beta_{3}X_{3} + \ldots + e
$$
* here $e$ means error, because we don't assume that any regression equation will perfectly predict $Y$
* other names for dependent variable are: outcome, response, explained
* other names for independent variable are: regressors, predictor, explantory

## Populations and Samples

* **population** - consists of all people or other entities that researchers want to study if they had infinite resources
* **sample** - subset of the population

### Nonprobability Sampling

* volunteer sample - put out an ad and get respondents; used when it would be difficult to sample from the population (ex. illegal drug users)
* convenience sample - interview 50 people shopping at a mall
* quota - shopping mall example but with added proportion demands such as 25 males and 25 females

### Probability Sampling

* simple random sampling - all samples of a given size have an equal probability of being selected
* systematic sampling - select every nth member
* complex random samples
    * stratified - population grouped based on strata or characteristics
    * cluster - preexisting groups (ex. geographic region); can be combined with sampling proportion to size
    
## Central Limit Theorem
states that the sampling distribution of the sample mean approximates the normal distribution, regardless of the distribution of the population if the sample size is sufficently large
$$
\overline{X} \dot\sim N(\mu \frac{\sigma^{2}}{n})
$$

## Hypothesis Testing

1. develop a research hypothesis that can be tested mathematically
2. formally state the null and alternative hypothesis
3. decide on an appropriate statistical test
4. make a decision based on the results

### Example
evaluating a new medication to treat high blood pressure (hypertension). The manufacturer wants to establish that it works better than currently available treatments for the same condition.

1. "hypertensive patients treated with new drug X will show greater lowering of their blood pressure than hypertensive patients treated with currently avalable drug Y"

2. stating the null and alternative hypothesis
$$H_{0}: \mu_{1} \leq \mu_{2}$$
$$H_{A}: \mu_{1} > \mu_{2}$$
$\mu_{1}$ = mean lowering of blood pressure in the group treated with drug X
$\mu_{2}$ = mean lowering of blood pressure in the group treated with drug Y
* $H_{0}$, or the null hypothesis states that drug X is no improvement over drug Y because the lowering of bp achieved by drug X os less than or equal to that achieved by drug Y
* $H_{1}$, or the alternative hypothesis states that drug x is more effective than standard treatment because patients treated with drug X show more lowring of bp than patients treated with drug Y.
* in this example the alt hypothesis is single-tailed because we state that drug X must achieve greater lowering of bp than drug Y. if we just wanted to claim that drug X and drug Y results differed, we could make a two-tailed claim and write it as:
$$H_{0}: \mu_{1} = \mu_{2}$$
$$H_{A}: \mu_{1} \neq \mu_{2}$$

after data is collected and statistics are calculated, we can make one of two decisions:
* reject the null hypothesis
    * we reject the null hypothesis when our statistical test shows that not only are their differences but that those differences are *statistically significant*, i.e. probablly not due to chance
* fail to reject the null hypothesis

### Error

* **type I error** - *alpha* or $\alpha$; when the null hypothesis is **true** but is rejected in a study
* **type II error** - *alpha* or $\beta$; when the null hypothesis is **false** but is not rejected in a study

the level of acceptability of type I error is traditionally set at 0.05, or that there is a 5% chance that we will reject the null hypothesis when we should fail to reject it.

## Confidence Intervals

* **point estimate** - single number; ex. mean
* **interval estimate** - range of numbers
* **confidence interval** - range between two values that represent the upper and lower confidence limits
* **confidence coefficient** - $(1 - \alpha)$, thus if $\alpha = 0.05$ the confidence coefficient is 0.95 or 95%


## p-values
the probability that results at least as extreme as those obtained in an analysis of sample data are due to chance.

### Example

flipping a coin that we believe to be fair ($P(H) = P(T) = 0.5$). our guess is that we get heads 50% of the time (5/10 trials).we actually get 80% (8/10 traisl). what is the the p-value of this result, that is, how likley is it that a coin with a probability of 0.5 for a single trial would produce 8 heads in 10 trials?

## Z-Statistic

analogus to the z-score, but instead of asking what the probability of a particular score is we are asking for the probability of a praticular sample mean.

$$
Z = \frac{\overline{x} - \mu}{\frac{\sigma}{\sqrt{n} }}
$$
$\overline{x}$ is the sample mean
$\mu$ is the population mean
$\sigma$ is the population standard deviation
$n$ is the sample size

* denominator is called the standard error of the mean (SEM)

## Data Transformations

* **parametric statistics** - make assumptions about the distrbution of the population from which the same was drawn
* **nonparametric statistics** - make few or assumptions about the distrbution of the population from which the same was drawn
* **transformation** - coerce the data to make sure that the assumptions of the desired parametric statistical procedure are met

* can use log or sqrt to transform
* statistics used are:
    * anderson-darling
    * kolmogorov-smirnov

* **data transformation is not a guaranteed solution to a distribution problem**
